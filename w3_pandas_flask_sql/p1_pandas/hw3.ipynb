При использовании функции .apply к датафрейму вдоль одной из осей даёт ответ в  виде Series обьекта. Однако данный ответ является отдельной от исходного датафрейма сущностью и результат простого применения функции .apply просто выдаётся в виде ответа, но не сохранятеся в структуре исходного датафрейма. Для сохранения изменений, получившихся в результате применения  функции .apply требуется в явном виде указать обьект, в котором будет проиведено сохранение резуьтата вычислений.
Изменения в изначальной таблице не были произведены, так как при выполнении лямбда - функции не было указания, каким столбцам таблицы должны соответсвовать результаты операции

```python
df[['Mjob', 'Fjob']] = df[['Mjob', 'Fjob']] .apply(lambda x: x.astype(str).str.upper())
df.head()
```

## Есть ли связь между потреблением алкоголя и академической успеваемостью студентов? Аргументируйте ваш ответ.

```python
df_alc = df[['Dalc','Walc','G1','G2','G3']]
```

### создание единого параметра уровня потребления алкоголя alc = int((Dalc' * 5 + Walc * 2)/7).

```python
df_alc['alc'] = df_alc.apply(lambda row: int((row['Dalc'] * 5 + row['Walc'] * 2)/7) , axis =1)
df_alc
```

### Исследование распределений успеваемости в зависимости от потребления алкоголя

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Data
df_viol = df_alc[['G3', 'alc']]

# Draw Plot
plt.figure(figsize=(13,10), dpi= 80)
sns.violinplot(x='alc', y='G3', data=df_viol, scale='width', inner='quartile')

# Decoration
plt.title('Violin Plot  уровень упеваемости при экзамене G3 в зависимости от уровня потребления алкоголя', 
          fontsize=22)
plt.show()
```

**С увеличением значения  параметра alc распределение успеваемости остается нормальным**


### Исследование зависимости медианы  успеваемости  от потребления алкоголя.

При работе с порядковыми параметрами лучше проводить исследование медиан, а не матожиданий.

```python
df_alc[['alc','G1','G2','G3']].groupby("alc").median()
```

**Медиана зачения успеваемости в зависимости от потребления алкоголя практически меняется (статистическая значимость различий не проводилась)**


### Исследование корреляционных зависимостей результатов аттестаций от потребления алкоголя 

Для порядковых параметров[ возможно использовать корреляцию Спирмена](https://en.wikipedia.org/wiki/Rank_correlation) для определения линейной зависимости

```python
plt.figure(figsize=(10,10), dpi= 80)
ax = sns.heatmap(df_alc.corr('spearman'), 
                 xticklabels=df_alc.corr('spearman').columns, 
                 yticklabels=df_alc.corr('spearman').columns, 
                 cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram', fontsize=22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
res = ax.get_figure()
```

**Корреляционной зависимости между группами 'alc', 'Dalc','Walc'  и 'G1','G2','G3' не выявлено.**
